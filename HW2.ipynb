1.  Write a python program to list all the odd numbers from 1 to 100<br><br>

```python
for i in range(100):
    if(i % 2 == 1):
        print(i)
```

2.  Write a python program to calculate all prime numbers between 1 and 100<br><br>

```python
for Number in range (1, 101):
    count = 0
    for i in range(2, (Number//2 + 1)):
        if(Number % i == 0):
            count = count + 1
            break

    if (count == 0 and Number != 1):
        print(Number)
```

4.  Write a lambda function to calculate the modulo of a number with a modulus of 2<br><br>

```python
modulo = lambda x: x % 2
```

5.  Starting with the lambda function in #3, write a map lambda function to calculate the iterable list x = [2, 6, 8, 10, 11, 4, 12, 7, 13, 17, 0, 3, 21]<br><br>

```python
x = (2, 6, 8, 10, 11, 4, 12, 7, 13, 17, 0, 3, 21)
y = list(map(modulo, x))
print(y)
```
