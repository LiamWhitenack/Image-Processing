## Complex Numbers

The Fourier transform inherently uses complex numbers in the form as $x + \imath y$<br>

Another representation of a complex number is $r e^{\imath \theta}$ <br>

The Complex Transformation operator converts all values in an image ${\bf d}[\vec x]$ from the linear representation to the polar representation as in

${\bf b}[\vec y] = P {\bf a}[\vec x]$

Each value in ${\bf a}[\vec x]$ is in $x+\imath y$ notation and each element in ${\bf b}[\vec x]$ is in $r^{\imath \theta}$ notation. The inverse operation is

${\bf a}[\vec x] = P^{-1} {\bf b}[\vec y]$

<br><br><br><br>


## Theory

The original theory was developed long before the onset of digital data and so it considers an analog signal $f(x)$

The Fourier transform is described as

$F(\omega) = \int_{-\infty} ^\infty f(x) ~ e^{-\imath \omega x} ~ dx$

where $\omega$ represents the frequency coordinates and $\imath=\sqrt{-1}$ 

Recall that $e^{\imath \theta} = \cos \theta + \imath \sin \theta$

and so the integrand is the multiplication of wave functions with $f(x)$

This function is basically the inner product (or projection) of $f(x)$ onto wave functions, and this will indicate the amount of each wave present in the original function

The magnitude of $F(\omega)$ is the magnitude of the $\omega$ frequency that is in $f(x)$ 

The inverse Fourier transform\index{inverse Fourier transform}\index{Fourier transform!inverse} is

$f^\prime (x) =\frac{1}{2 \pi} \int_{-\infty} ^\infty F(\omega) ~ e^{\imath \omega x} ~ d\omega$

There is no loss of information and so $f^\prime (x) = f(x)$

<br><br><br><br>

Any reasonable function can be expressed as a (infinite) linear combination of sines and cosines!

$\mathrm{F}(\mathrm{t})=\mathrm{a}_{0}+\mathrm{a}_{1} \cos (\omega \mathrm{t})+\mathrm{b}_{1} \sin (\omega \mathrm{t})+a_{2} \cos (2 \omega t)+b_{2} \sin (2 \omega t)+\ldots$

$=\sum_{n=0}^{\infty}\left(a_{n} \cos (n \omega t)+b_{n} \sin (n \omega t)\right)$

where, $F(t)$ is a periodic function with $\omega=\frac{2 \pi}{T}$

$F(t)$ must satisfy certain other conditions
* finite number of discontinuities within T
* finite average within T
* finite number of minima and maxima

The coefficients are:

$F(t)=\sum_{n=0}^{\infty}\left(a_{n} \cos (n \omega t)+b_{n} \sin (n \omega t)\right.$

where, $a_{k}=\frac{2}{T} \int_{0}^{T} f(t) \cos (k \omega t) d t \quad b_{k}=\frac{2}{T} \int_{0}^{T} f(t) \sin (k \omega t) d t$

## Digital Fourier Transform

Data is digital and therefore it is necessary to review the digital Fourier transform

To distinguish an analog signal from a digital signal, the input has a slightly different form with the use of square brackets

The <b>DFT</b> (digital Fourier transform) of a signal $f[n]$

$F[\omega] = \sum_{n=1} ^{N-1} f[n] ~ e^{-2 \pi \imath n \omega /N}$

where $N$ is the number of elements in the signal

The inverse <b>DFT</b> is,

$f^\prime[n] = \frac{1}{N} \sum_{\omega=1} ^{N-1} F[\omega] ~ e^{2 \pi \imath n \omega /N}$

The philosophy is the same in that the Fourier transform is the projection of the original signal onto a set of frequencies, and the inverse transform is the weighted linear combination of those frequencies

For two dimensional signals the transforms are,

$F[u,v] = \sum_{n=0}^{N-1} \sum_{m=0} ^{M-1} f[x,y] e^{-\imath 2 \pi (ux/M + vy/N)}$

and the inverse is

$f(x,y) = \frac{1}{MN} \sum_{n=0}^{N-1} \sum_{m=0} ^{M-1} F[u,v] e^{\imath 2 \pi (ux/M + vy/N)}$

The operator for a Fourier transform is $\mathfrak{F}$ and the inverse is $\mathfrak{F}^{-1}$

<br><br><br><br><br><br><br><br><br><br><br><br>


## FFT in Python

The digital Fourier transform contains a large amount of repetitive computations

The Cooley-Tukey algorithm provides an efficient method of performing the Fourier transforms by eliminating these massive repetitive computations

This fast Fourier transform (FFT) is the algorithm of choice that is used in computer programs

The scipy module contains a package named fftpack with several Fourier transform functions

The function <b>fft</b> computes the forward Fourier transform and the function <b>ifft</b> computes the inverse transform

Two dimensional transforms use the functions <b>fft2</b> and <b>ifft2</b>

Higher dimensional arrays are transformed by the functions <b>fftn</b> and <b>ifftn</b>

<br><br>

Remember the Plop operator will place the an image inside of a larger frame and is written as,

${\bf b}[\vec x] = U_{\vec w} {\bf a}[\vec x]$

where $\vec w$ is the frame size of ${\bf b}[\vec x]$

Many applications would prefer that the frame size of the Fourier transform be the same as the frame size of the input

One manner of performing this is to plop the image into a larger frame, perform the Fourier transform, and then cut out the middle portion of the transform, as in,

${\bf b}[\vec \omega] = \Box_{\vec v_1, \vec v_2} \mathfrak{F} U_{\vec w} {\bf a}[\vec x]$

where $\vec v_1$ and $\vec v_2$ are the upper left and lower right locations of the portion to be cut out

The vectors $\vec v_1$ and $\vec v_2$ are designed to extract the center portion of the Fourier transform, and to make ${\bf b}[\vec x]$ is the same frame size as ${\bf a}[\vec x]$

<br><br>

```python
#  simple example of an FFT
import numpy as np
from scipy.fft import fft, ifft
np.set_printoptions(precision=2)

x = ([5.0])
print('\nOur input scalar:\n', x)

y = fft(x)
print('\nFFT of a scalar:\n', np.round(y, 2))

x = np.array([1.0, 2.0, 1.0, -1.0, 1.5])
print('\nOur input vector:\n', x)

y = fft(x)
print('\nFFT of our vector:\n', np.round(y, 2))
```

```python
print('\nOur input vector:\n', x)

yinv = ifft(y)
print('\nInverse FFT of our FFT:\n', yinv)
```

## Properties of a Fourier Transform

The DC term is $F[0]$ which is also the sum of the input signal, and it is called the bias term<br>

This is the average height of the signal<br>

In the case of the above figure, the average height is 0 and thus the first term in the Fourier transform is zero<br> 

If the entire input signal was raised by a value of $h$ then $F[0] = N h$ (where $N$ is the number of elements) but the other terms are not changed<br>

Thus, for an image, the first term of the FFT is

$\mathbf{b}[(0,0)]=\sum_{\vec{x}} \mathbf{a}[\vec{x}]$


```python
#  DC (the sum of the input signal, it is called the bias term) term example
import numpy as np
import scipy.fftpack as ft

np.random.seed(21434)
mga = np.random.ranf((32,32))
mgb = ft.fft2(mga)
print('\nSum of input signal:', round(mga.sum(),2))

print('\nDC Term:', np.round(mgb[0,0],2))

print('\nDC Term:', np.round(mgb[1,1],2))

print('\nDC Term:', np.round(mgb,2))
```

## Conservation of Energy

Energy is conserved in a Fourier transform for it is merely a transformation<br>

Energy is not added or subtracted in the process<br>

This must be true or the inverse transform will not retrieve the original signal<br>

Parseval's theorem relates the energy of the input space and the frequency space as the sum (or integral) of the square of a function is equal to the sum (or integral) of the square of its transform

$\sum_{\vec x} | {\bf a}[\vec x] |^2 = \frac{1}{N{\bf a}[\vec x]} \sum_{\vec x} | \mathfrak{F} {\bf a}[\vec x] |^2$


```python
x = 1 + 2j
y = np.conjugate(x)
print(y)
print('\nSimple example of multiplying two imaginary numbers:', x*y)

#  conservation of energy
a = (mga*mga.conjugate()).sum()
print('\nLeft side of equation:', np.round(a,2))

b = (mgb*mgb.conjugate()).sum()/(32*32)
print('\nRight side of equation:', np.round(b,2))
```

## Replication

The inverse transform returns the original signal without loss of information 

If ${\bf b}[\vec \omega] = \mathfrak{F}{\bf a}[\vec x]$ and ${\bf c}[\vec x] = \mathfrak{F}^{-1} {\bf b}[\vec \omega]$, then ${\bf a}[\vec x] = {\bf c}[\vec x]$ is also true

Comparing two arrays in Python can be accomplished with the <b>allclose</b> function

This will compare the arrays, element by element, and consider them to be the same if the values are within tight precision which is set by default to $1.0 \times 10^{-5}$

The return of <b>True</b> indicates that the two signals are the same


```python
# replication
mgc = ft.ifft2(mgb)
print('\nAre the two arrays the same:', np.allclose(mga, mgc))

print('\n', mga-mgc)
```

## Addition

The Fourier transform of the sum of two functions is the same as the sum of the Fourier transform of the functions.  Consider a case in which the function $f(x)$ can be written as the summation of two functions,
$f(x) = g(x) + h(x)$

The Fourier transform of $f(x)$ is then,

$F[\omega] = \int (g(x) + h(x)) e^{-\imath \omega x} ~dx =  \int g(x) e^{-\imath \omega x} ~dx +  \int h(x)  e^{-\imath \omega x} ~dx$

These two integrations are just individual Fourier transforms and so,

$F[\omega] = G[\omega] + H[\omega]$

In operator notation, if 

${\bf a}[\vec x] = {\bf b}[\vec x] + {\bf c}[\vec x]$

then

$\mathfrak{F} {\bf a}[\vec x] = \mathfrak{F} {\bf b}[\vec x] + \mathfrak{F} {\bf c}[\vec x]$

<br><br><br><br><br><br><br><br><br><br><br><br>


## Shift

A lateral shift of a signal creates a phase shift in Fourier space. If ${\bf g}[\vec \omega] = \mathfrak{F} {\bf f}[\vec x]$ then 

$e^{- \imath 2 \pi \vec \omega \cdot \vec v} {\bf g}[\vec \omega] = \mathfrak{F} D_{\vec v} {\bf f}[\vec x]$


```python
#  shift
import numpy as np
import imageio
import scipy.misc as sm
import scipy.fftpack as ft

url = 'https://raw.githubusercontent.com/joefoxva1/CDS468/master/Lecture2_3_4/bird_js.jpg'
mga = imageio.imread(url, as_gray=True).astype(np.uint8)

#  left side
V,H = mga.shape
shift = 2
mgg = ft.fft2(mga)
w = np.indices((V,H)).transpose((1,2,0))
v = np.array((0,shift) )/np.array((V,H))
wv = w.dot( v )
phase = np.exp(-1j*2*np.pi * wv)
fmgg = phase * mgg

#  right side
mgf = np.zeros(mga.shape)
mgf[:,:shift] = mga[:,-shift:]+0
mgf[:,shift:] = mga[:,:-shift]+0
fmgf = ft.fft2(mgf)

print('\nAre the two arrays the same:', np.allclose(fmgg,fmgf,atol=0.05))
```

## Scale

The scale property relates a change of input space to the change of scale in the output space

Basically, if the width of a signal of the input decreases then the width of the Fourier pattern expands  

The change in scale of a signal by a factor $a$ is represented by $f(ax)$

The Fourier transform is

$F(\omega) = \int f(ax)~ e^{-\imath \omega x} ~dx$

Using $u \equiv ax$, then $dx = \frac{1}{a} du$, and the transform becomes

$F(\omega) = \frac{1}{a} \int f(u)~e^{-\imath \omega u/a} ~ du = \frac{F(\omega/a)}{a}$

If $a>1$ then the scale of the input space increases while the scale in the Fourier space decreases

In operator notation, if ${\bf b}[\vec \omega] = \mathfrak{F}{\bf a}[\vec x]$, then the scaling property is,

$\frac{{\bf b}[\vec \omega/a]}{a} = S_a {\bf a}[\vec x]$

## Displaying the Transform

There are three issues that need to be addressed before it is possible to create a viewable Fourier transform image<br>

The first is that the Fourier transform contains complex values<br>

The second is that the DC term is generally several orders of magnitude greater than the other values<br>

The third is that the low frequencies are at the corners and the preference is that they be in the middle of the frame<br>

<br><br>

A common method of displaying a Fourier transform is

${\bf c}[\vec \omega] = X L ( | {\bf b}[\vec \omega] | + 1 )$

The $X$ operator is the quadrant swap<br>

The image is divided into four quadrants<br>

The upper left and lower right quadrants exchange positions, and the other two quadrants exchange positions<br>

The Python function for this is  scipy.fftpack.fftshift<br>

Since a pixel on the computer screen has only 256 levels of gray, any value that is less than 1/256 of the largest value will appear as black<br>

The solution is to display the log values, and that operator is $L$. The absolute value converts the complex values to real values, and the addition of 1 will ensure that all log values are positive<br>

## Frequency Bands

Fingerprints are unique in that most of the information is stored in frequencies instead of intensities<br>

Furthermore, the spacing between the ridges is about the same throughout the image<br>

Thus, these images are ideal for exploring frequency band filtering<br>

A whorl print pattern is shown below. The image displayed below (b) is the Fourier transform display image<br>

In the center is a bright spot that is associated with the DC term. About that center is a strong primary ring and a weaker secondary ring

<br>

<br>

The radius of the primary ring is directly related to the spacing between ridges in the fingerprint<br>

The ring has a width since the spacing between the ridges is not exactly the same throughout the fingerprint<br>

Most of the information about the fingerprint is contained within this ring<br>

Most of the fingerprint information is contained in the bright ring within in the Fourier image<br>

To support this statement, the rest of the frequencies are removed, and then image is transformed back to the original image space 

The process is,

${\bf b}[\vec x] = \Box_{\vec v_1, \vec v_2} | \mathfrak{F}^{-1} X \left ( {\bf m}[\vec \omega] \times X \mathfrak{F} U_{\vec w} {\bf a}[\vec x] \right) |$

where ${\bf m}[\vec \omega] = {\bf o}_{\vec w; r_2} - {\bf o}_{\vec w; r_1}$<br>

The $\vec w$ is the size of the frame, $r_2$ is a radius that is slightly bigger than the outer part of the white ring, and $r_1$ is a radius that is slightly smaller than the inner part of the white ring<br>

The ${\bf m}[\vec \omega]$ is an annular ring mask that encompasses the white ring in the Fourier image<br>

To perform the Fourier transform the image is placed inside of a frame using $U_{\vec w}$ were $\vec w = (1024,1024)$<br>

The image is processed but the final operator reduces the frame size back to that of the original image using $\Box_{\vec v_1, \vec v_2}$, where $\vec v_1 = (140,160)$ and $\vec v_2 = (885,863)$<br>

The multiplication of the mask by the swapped Fourier transform of the image will block out all of the frequencies that are not inside of this annular ring<br>

It is removing the frequencies that are not in, or very near, the white band. The quadrants are swapped again before applying the inverse transform<br>

The absolute value of the complex values are obtained before cutting the frame size<br>


```python
def Circle(size, loc, rad):
    """frame, center, radius size is (v,h) of size of array loc is (v,h) of circle location
    rad is integer of radius returns array with a solid circle"""
    b1, b2 = np.indices(size)
    b1, b2 = b1-loc[0], b2-loc[1]
    mask = b1 * b1 + b2 * b2
    mask = (mask <= rad*rad).astype(int)
    return mask

def Plop(data, VH, back=0):
    # vmax, hmax are size of frame
    ans = np.zeros( VH, float ) + back
    vmax, hmax = VH
    # plops the data in the center
    # get center of blob
    V,H = data.shape
    vctr, hctr = V//2, H//2
    vactr, hactr = vmax//2, hmax//2
    # compute the limits for the answ
    valo = vactr - vctr
    if valo<0: valo = 0
    vahi = vactr + vctr
    if vahi>=vmax: vahi = vmax
    halo = hactr - hctr
    if halo<0: halo = 0
    hahi = hactr + hctr
    if hahi>=hmax: hahi = hmax
    # compute limits of incoming
    vblo = vctr - vactr
    if vblo<=0: vblo = 0
    vbhi = vctr + vactr
    if vbhi>=V: vbhi= V
    hblo = hctr - hactr
    if hblo<=0: hblo = 0
    hbhi = hctr + hactr
    if hbhi>=H: hbhi = H
    if vahi-valo != vbhi-vblo:
        vbhi = vblo+vahi-valo
    if hahi-halo != hbhi-hblo:
        hbhi = hblo+hahi-halo
    ans[valo:vahi, halo:hahi] = data[vblo:vbhi, hblo:hbhi] + 0
    return ans

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import imageio
import scipy.fftpack as ft

url = 'https://raw.githubusercontent.com/joefoxva1/CDS468/master/Lectures10_11_12/fing.png'
adata = 256 - imageio.imread(url, as_gray=True).astype(np.uint8)

plt.title('Original Image')
plt.imshow(adata, cmap='gray')
plt.show()

V, H = 1024, 1024
adata = Plop(adata, (V,H))
plt.title('Original Image Plop in Larger Window')
plt.imshow(adata, cmap='gray')
plt.show()

mask = Circle((V,H),(V/2,H/2), 60)
mask -= Circle((V,H),(V/2,H/2), 20)
plt.title('Mask being applied')
plt.imshow(mask, cmap='gray')
plt.show()

temp1 = ft.fftshift(ft.fft2(adata))
bdata = abs(ft.ifft2(ft.fftshift(mask*temp1)))
plt.title('Resulting Image')
plt.imshow(bdata[140:885,160:863], cmap='gray')
plt.show()
```

##  Windowing

The Fourier theory assumes that the input signal is repetitive<br>

Thus, the input signal can be considered as concatenations of the image in both the vertical and horizontal directions<br>

The presence of vertical and horizontal lines arise from the discontinuity  between the top and bottom rows as well as the left-most and right-most columns<br>

The Fourier transform of an image that fills the frame can produce the cross hairs because of the discontinuities along the edges of the frame<br>

One solution to this situation is to apply a window to the original image<br>

A window is a filter that blocks some of the pixels and allows others to maintain their original value<br>

In this case, the window will block several rows of pixels along each edge of the frame<br>

Given the frame size $\vec w = (V,H)$, the filter is ${\bf r}_{\vec w; \vec v_1, \vec v_2}$ where $\vec v_1 = (z,z)$ and $\vec v_2 = (V-z,H-z)$<br>

This is a window with a sharp edge will still cause the same issue with the cross hairs, so a smoothing operator\index{Smooth operator} is applied as in $\mathcal{S}_m {\bf r}_{\vec w; \vec v_1, \vec v_2}$.<br>

Thus, windowing is performed by a simple mask that is a rectangle with smoothed edges. Given the grayscale input image, ${\bf a}[\vec x]$, the entire process is described by,

${\bf b}[\vec x] = (X \mathfrak{F} ( {\bf a}[\vec x]) \times  \mathcal{S}_m {\bf r}_{\vec w; \vec v_1, \vec v_2} [\vec x]$

where $m=5$, $\vec w = (484,512)$, $\vec v_1 = (20,20)$, and $\vec v_2=(464,492)$ 


```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import imageio
import scipy.fftpack as ft
import scipy.ndimage as nd

url = 'https://raw.githubusercontent.com/joefoxva1/CDS468/master/Lectures10_11_12/fing.png'
adata = 256 - imageio.imread(url, as_gray=True).astype(np.uint8)

plt.title('Original Image')
plt.imshow(adata, cmap='gray')
plt.show()

V, H = [745, 703]
mask = np.zeros((V,H))
mask[250:-20,250:-20] = 1
mask = nd.gaussian_filter(mask, 5)
bdata = ft.fftshift(ft.fft2(adata*mask))
cdata = bdata.real

plt.title('The mask')
plt.imshow(mask, cmap='gray')
plt.show()

cdata = abs(ft.ifft2(ft.fftshift(mask*bdata)))
plt.title('Image after applying the mask')
plt.imshow(cdata, cmap='gray')
plt.show()
```

<br><br><br><br><br><br><br><br><br><br><br><br>

There are various types of windows that have been explored which treat the smoothing of the edges with slightly different functions

The Hamming window for one dimension is defined as,

$w(n) = 0.54 - 0.46 \cos\left(\frac{2\pi{n}}{M-1}\right) \qquad 0 \leq n \leq M-1$

where $n$ is the pixel location

Another alternative is a Hanning window defined by,

$w(n) = 0.5 - 0.5 \cos\left(\frac{2\pi{n}}{M-1}\right) \qquad 0 \leq n \leq M-1$

The scipy.signal module offers several such windows 

While analog implementations of these windows have differing results, the digital implementation is not as sensitive<br>

<br><br><br>

Usually, there are only a few pixels within the region that the window operation affects, and so most windows produce very similar results<br>

For most applications the use of a window is important but the type of window is not<br>

These Python functions create windows in one-dimension and it would be up to the user to create a two-dimensional window<br> 

Then there is the task of resizing and locating the window for a particular function<br>

An alternative is the KaiserMask which is a lengthy function and therefore relegated to KaiserMask<br>

This creates a two-dimensional matrix which contains a circular region with edges decaying as a Kaiser window

The operator notation is,

${\bf a}[\vec x] = {\bf k}_{\vec w; r_1, r_2} [\vec x]$

<br><br><br>

In the code below, the first is the size of the frame and the second is the center of the masking region<br>

The final two arguments are the radii

In this example the mask is a circle with a center at (128,128)<br>

The pixels inside of a radius of 50 have a value of 1<br>

The pixels beyond a radius of 50 have a value of 0<br>

The pixels in between the two radii decay from 1 to 0<br>

This mask is useful in both the image and frequency spaces<br>

This mask can be used as a window and replace $ \mathcal{S}_m {\bf r}_{\vec w; \vec v_1, \vec v_2} [\vec x]$ in the equation above<br>


```python
def KaiserMask(shape, center, r1, r2):
    # returns a mask that you can use over and over again.
    di, dj = center  # location of the center of the window
    v,h = shape
    theta = 2. * np.pi
    Iot = 1.0 + theta/4. + 2.*theta/64. + 3.*theta/2304.
    
    # compute radii
    vindex = np.multiply.outer(np.arange(v), np.ones(h))
    hindex = np.multiply.outer(np.ones(v), np.arange(h))
    a = (di-vindex).astype(float)
    b = dj-hindex
    r = np.sqrt(a*a + b*b)
    del a,b
    
    # inside r1, and outside r2 are easy
    mask = np.zeros(shape, float)
    mask = (r < r1).astype(int)
    
    # work on the ring between
    b = np.logical_and((r<r2), (r>r1))
    m = (r-r1)/(r2-r1)
    m = m*b
    a = theta * np.sqrt( 1.-m*m)
    a = 1.0 + a/4.0 + 2.0*a/64.0 + 3.0*a/2304.0
    a = a / Iot
    a = a * (r< r2)
    a = a * (r>=r1).astype(int)
    mask = mask + a
    return mask

mask = KaiserMask((256,256), (128,128), 0, 50)

plt.title('Kaiser Mask')
plt.imshow(mask, cmap='gray')
plt.show()
```

## Summary

The Fourier transform converts an image from its spatial representation to its frequency representation<br>

The transform consists of complex values over a wide range of magnitudes<br>

Thus, the display of the information is often the log of the absolute values<br>

The inverse Fourier transform returns the information back to the image space<br>

These transformations do not lose information and so the consecutive application of the forward and inverse transforms will return the original image<br>

However, the transform does allow access to the frequencies, and thus many applications will alter the frequencies in some manner before returning the information back to image space<br>

Thus, the image is modified and usually providing the user with better access to the pertinent information
