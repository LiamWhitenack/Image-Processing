## Artificial Neural Networks

<br>

<b>The brain contains a largely connected network of neurons</b><br>

<b>Output of a single neuron may be the input to thousands of other neurons</b><br>

<b>Learning occurs by repeatedly activating certain neural connections over others</b><br>

<b>Connections then are reinforced</b><br><br>

<div> <img src="https://github.com/joefoxva1/GGS_590_DL/blob/master/Lecture6_7/NN1a.png?raw=true" width="400"/img></div>

<b>Reinforcement makes them more likely to product the desired outcome given a specific input</b><br>

<b>The learning involves feedback</b><br>

<b>When the desired outcome occurs, the neural connections causing the outcome becomes strengthened</b><br>

<b>Artificial Neural Networks (ANN) attempt to simplify and mimic this brain behavior</b><br>

<b>ANNs can be train in a supervised or unsupervised manner</b><br>

<b>Supervised ANN – network is trained by providing matched input and output data samples:  Example is SPAM filter</b><br>

<b>Unsupervised ANN – attempt to get the ANN to understand the structure of the input data on its own</b><br>

<br><br><br><br>

## Structure of an ANN

<b>Neuron is simulated by an activation function</b><br>

<b>Activation function must have a switch on; once the input is greater than a certain value, the output should change state (0 to 1 or -1 to 1, etc.)</b><br>

<b>Simulates the turning on of a neuron</b><br>

<b>Common activation function is the sigmoid function</b><br><br>

<div> <img src="https://github.com/joefoxva1/GGS_590_DL/blob/master/Lecture6_7/NN2.png?raw=true" width="300"/img></div>

<b>Example:  </b>
$f(x)=\frac{1}{1+e^{-z}}$

<br><br><br><br>

```python
#  *****  Sigmoid function
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np

x = np.arange(-8, 8, 0.1)       #  range of x values
f = 1/(1 + np.exp(-x))          #  sigmoid function

#  plot the function
plt.title('Sigmoid Function')
plt.plot(x,f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

### Weight Adjustment

```python
#  *****  weight adjustment
%matplotlib notebook
import matplotlib.pylab as plt
import numpy as np

x = np.arange(-8, 8, 0.1)       #  range of x values

w1 = 0.5
w2 = 1.0
w3 = 2.0
l1 = 'w = 0.5'
l2 = 'w = 1.0'
l3 = 'w = 2.0'
for w, l in [(w1, l1), (w2, l2), (w3, l3)]:
    f = 1/(1 + np.exp(-x*w))          #  sigmoid function
    plt.plot(x, f, label=l)

plt.title('Sigmoid Function with Varying Weights')
plt.xlabel('x')
plt.ylabel('h_w(x)')
plt.legend(loc=2)
plt.show()
```

## Bias

<br>

<b>Changing the weight changed the slope of the sigmoid function</b><br>

<b>Useful in modeling different strengths between the input and output variables</b><br>

<b>What if we want to change the output when x is greater than 1?  Add bias</b><br><br>

<div> <img src="https://github.com/joefoxva1/GGS_590_DL/blob/master/Lecture6_7/NN7.png?raw=true" width="300"/img></div>

<b>For Example: Add bias</b><br>

* b1 = -8.0

* b2 = 0.0
* b3 = 8.0

<br><br>
<div> <img src="https://github.com/joefoxva1/GGS_590_DL/blob/master/Lecture6_7/NN8.png?raw=true" width="500"/img></div>

<br><br><br><br>

```python
#  *****  bias
%matplotlib inline
import matplotlib.pylab as plt
import numpy as np

x = np.arange(-8, 8, 0.1)       #  range of x values

w = 5.0
b1 = -8.0
b2 = 0.0
b3 = 8.0
l1 = 'b1 = -8.0'
l2 = 'b2 = 0.0'
l3 = 'b3 = 8.0'
for b, l in [(b1, l1), (b2, l2), (b3, l3)]:
    f = 1/(1 + np.exp(-x*w+b))          #  sigmoid function with bias
    plt.plot(x, f, label=l)

plt.title('Sigmoid Function with Varying Bias')
plt.xlabel('x')
plt.ylabel('h_w(x)')
plt.legend(loc=2)
plt.show()  
```

<br><br>

## Effect of Bias Adjustments

<br>

<b>w was increased to simulate a more defined turn-on function</b><br>

<b>By varying the bias weight b, the node activation point changes</b><br>

<b>By adding a bias term, you are able to simulate and “if” function (useful for conditional relationships)</b><br>

```python
#  feed forward network - brute force
import numpy as np

#  layer 1 weights
w1 = np.array([[0.2, 0.2, 0.2], [0.4, 0.4, 0.4], [0.6, 0.6, 0.6]])

# layer 2 weights
w2 = np.zeros((1,3))
w2[0,:] = np.array([0.5, 0.5, 0.5])

# bias values
b1 = np.array([0.8, 0.8, 0.8])
b2 = ([0.2])

#  activation function - sigmoid
def f(x):
    return 1/(1+np.exp(-x))

def simple_looped_nn_calc(n_layers, x, w, b):
    for l in range(n_layers-1):
        #  if the first layer then the input array will be the x vector
        #  if not the first layer, the input is the output of the previous layer
        if l == 0:
            node_in = x
        else:
            node_in = h
    
        #  setup the output array for nodes in layer +1
        h = np.zeros((w[l].shape[0],))
        
        #  loop through the rows of the shape array
        for i in range(w[l].shape[0]):
            f_sum = 0
            for j in range(w[l].shape[1]):  #  loop through all the columns
                f_sum += w[l][i][j] * node_in[j]
            
            f_sum += b[l][i]  #  add the bias
            h[i] = f(f_sum)
    return h

w = [w1, w2]
b = [b1, b2]

x = [1.5, 2.0, 30.0]  #  layer 1

q = simple_looped_nn_calc(3, x, w, b)
print('\nOutput of a simple feed forward NN', round(q[0],3))
```

```python
#  feed forward network - matrix way
import numpy as np

#  layer 1 weights
w1 = np.array([[0.2, 0.2, 0.2], [0.4, 0.4, 0.4], [0.6, 0.6, 0.6]])

# layer 2 weights
w2 = np.zeros((1,3))
w2[0,:] = np.array([0.5, 0.5, 0.5])

# bias values
b1 = np.array([0.8, 0.8, 0.8])
b2 = ([0.2])

#  activation function
def f(x):
    return 1/(1+np.exp(-x))

def matrix_nn_calc(n_layers, x, w, b):
    for l in range(n_layers-1):
        #  if the first layer then the input array will be the x vector
        #  if not the first layer, the input is the output of the previous layer
        if l == 0:
            node_in = x
        else:
            node_in = h
        
        z = w[l].dot(node_in) + b[l]
        h = f(z)
    return h

w = [w1, w2]
b = [b1, b2]

x = [1.5, 2.0, 3.0]

q = simple_looped_nn_calc(3, x, w, b)
print('\nOutput of a simple feed forward NN', round(q[0],3))
```

## Gradient Decent Redux

<br>

<b>Idea of supervised learning is to provide many input-output pairs of known data and vary the weights based on these samples so that the error expression is minimized</b><br>

<b>In training the network with these (x,y) pairs, the goal is to get the neural network better and better at predicting the correct y given x</b><br>

<b>This is performed by varying the weights so as to minimize the error</b><br><br>

<div> <img src="https://github.com/joefoxva1/GGS_590_DL/blob/master/Lecture6_7/NN17.png?raw=true" width="300"/img></div>

<b>Gradient descent method uses the gradient to make an informed step change in w to lead it towards the minimum of the error curve</b><br>

<b>In each step, the w value is updated according to:</b><br>

$\boldsymbol{w}_{\boldsymbol{n e w}}=\boldsymbol{w}_{\text {old}}-\boldsymbol{\alpha} * \boldsymbol{\delta}$ error<br>

<b>Gradient Decent Cost Function</b><br>

$J(x)=-\frac{1}{2 n} \sum_{i}\left(y_{i} \theta x_{i}-\theta x_{i}-\ln \left(1-e^{-\theta x}\right)\right)$<br>

<b>Gradient Direction of Travel</b><br>

$\frac{d}{d \theta} J(x)=-\frac{1}{2 n} \sum_{i}\left(x_{i}\left(y_{i}-\frac{1}{1+e^{-\theta x}}\right)\right.$<br>

<br><br>

## Gradient Decent – Example Code

Solve: $f(x)=x^{4}-3 x^{3}+2$<br>

Solve: $f(x)=e^{-x}+2 x^{3}+3 x+7$

<br><br>

```python
# example implementation gradient decent
x_old = 0     # does not matter
x_new = 6     # place to start
gamma = 0.01  #  step size
precision = 0.00001

def df(x):
#    y = 4 * x**3 - 9 * x**2
    y = -np.exp(-x) + 6*x**2 + 3 
    return y

while abs(x_new - x_old) > precision:
    x_old = x_new
    x_new += -gamma * df(x_old)

print('\nThe local minimum occurs at %f' % x_new)
```
